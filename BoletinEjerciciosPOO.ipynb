```python
# 3001 - Galleta
# Definir una clase Galleta que tiene dos atributos: nombre de la galleta y su forma. La clase Galleta debe de contener el contructor y el siguiente método:
# __init__(self,nombre): un contructor que permite inicializar los atributos de la clase Galleta.
# __hornear__(self): un método que permite mostrar dos frases en la consola:
# ✔ Esta galleta con chispas de chocolate ha sido horneada en forma de estrella
# ✔ Buen provecho.

class Galleta:
    def __init__(self, nombre):
        self.nombre = nombre
        self.forma = "Estrella"

    def __hornear__(self):
        print(f"Esta galleta {self.nombre} ha sido horneada en forma de {self.forma}")
        print("¡Buen provecho!")

galleta1 = Galleta("Tostarica")
galleta1.__hornear__()
```

```python
# 3002 - Libro
# Definir una clase Libro que tienes 3 atributos, título del libro, nombre del autor y precio del libro. La clase Libro debe contener el constructor y el siguiente método:
# __init__(self,titulo,autor,precio): constructor que permite inicializar los atributos de la clase Libro
# Mostrar_informaciones(self): permite mostrar la información relacionada con el libro.

class Libro:
    def __init__(self, titulo, autor, precio):
        self.titulo = titulo
        self.autor = autor
        self.precio = precio

    def Mostrar_informacion(self):
        print(f"Título: {self.titulo} Autor: {self.autor} Precio: {self.precio}€")

libro1 = Libro("El Paraíso Perdido", "John Karpenter", 20)

libro1.Mostrar_informacion()
```

```python
# 3003 - Nota
# Definir una clase Nota que tiene 2 atributos, nombre del estudiante y su nota. La clase Nota debe contener el constructor y el siguiente método:
# __init__(self,nota,nombreEstudiante): constructor que permite inicializar los atributos de la clase Nota
# ha_pasado(self): El método muestra en consola si el alumno está aprobado o suspenso. Si la nota del estudiante supera los 75 puntos entonces el alumno ha aprobado,
# de lo contrario está suspenso.
class Nota:
    def __init__(self, nota, nombreEstudiante):
        self.nota = nota
        self.nombreEstudiante = nombreEstudiante

    def ha_pasado(self):
        if self.nota >= 75:
            print(f"El alumno: {self.nombreEstudiante} ha aprobado!")
        else:
            print(f"El alumno: {self.nombreEstudiante} ha suspendido!")

alumno1 = Nota(85, "Pablo")
alumno1.ha_pasado()
alumno2 = Nota(60, "Juan")
alumno2.ha_pasado()
```

```python
# 3004 - Círculo
# Definir una clase Circulo que tiene 3 atributos: el radio y la posición de su centro O(x,y). La clase debe de contener el constructor y los siguientes métodos:
# __init__(self,x,y,r): constructor que permite inicializar los atributos de la clase Circulo
# area(self): calcula el área y la muestra
# perimetro(self): calcula el perímetro y lo muestra
# mostrar_propiedades(self): muestra los valores de los atributos por consola.
class Circulo:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def area(self):
        return (self.x * self.y) * (self.z * self.z)

    def perimetro(self):
        return (self.x * self.y) + (self.y * self.z)

    def mostrar_propiedades(self):
        print(f"Area: {self.area()}")
        print(f"Perimetro: {self.perimetro()}")

Circulo1 = Circulo(5, 10, 20)
Circulo1.mostrar_propiedades()
```

```python
# 3005 - Operación
# Definir una clase Operacion que tiene dos atributos: dos números enteros x e y. Esta clase nos permitirá realizar operaciones aritméticas entre con estos dos atributos.
# La clase Operación debe contener el constructor y los siguientes métodos:
# __init__(self,x,y): constructor que permite inicializar los atributos de la clase Operacion
# suma(self): realiza y muestra la suma de los atributos
# multiplicación(self): realiza y muestra la multiplicación
# division(self): realiza y muestra la división. Debe controla que no se puede dividir entre 0.
class Operacion:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def suma(self):
        print(f"Suma: {self.x + self.y}")

    def multiplicacion(self):
        print(f"Multiplicacion: {self.x * self.y}")

    def division(self):
        if self.y == 0:
            print(f"¡No se puede dividir por cero!")
        else:
            print(f"Division: {self.x / self.y}")

Operacion1 = Operacion(10, 2)
Operacion1.suma()
Operacion1.multiplicacion()
Operacion1.division()
```

```python
# 3006 - Persona
# Definir una clase Persona que tiene tres atributos: nombre, edad y género. La clase Persona debe contener el constructor y los siguientes métodos:
# __init__(self,nombre,edad,genero): constructor que permite inicializar los atributos de la clase Persona
# presentarse(self): el método retorna una cadena
# esAdulto(self): retorna True si la persona es igual o mayor de 18 años, false en caso contrario
class Persona:
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre
        self.edad = edad
        self.genero = genero

    def presentarse(self):
        print(f"Hola, mi nombre es {self.nombre}. Tengo {self.edad} de edad y soy {self.genero}")

    def esAdulto(self):
        if int(self.edad) >= 18:
            print(True)
        else:
            print(False)

Persona1 = Persona("Pablo", "21", "Masculino")
Persona1.presentarse()
Persona1.esAdulto()
Persona2 = Persona("Juan", "15", "Masculino")
Persona2.presentarse()
Persona2.esAdulto()
```

```python
# 3007 - Estudiante
# Definir una clase Estudiante que hereda de la clase Persona. La clase Estudiante debe contener el constructor y los siguientes métodos:
# __init__(self,nombre,edad,genero,nivel): constructor que permite inicializar los atributos de la clase Estudiante
# inscripcion(self,estudiantes_inscritos): recibe como parámetro una lista de estudiantes y añade el nuevo estudiante a la lista
class Estudiante(Persona):
    def __init__(self, nombre, edad, genero, nivel):
        super().__init__(nombre, edad, genero)
        self.nivel = nivel

    def inscripcion(self, estudiantes_inscritos):
        estudiantes_inscritos.append(self)
        print(f"Estudiante {self.nombre} inscrito correctamente.")

    def __str__(self):
        return f"{super().__str__()} {self.nivel}"

lista_estudiantes = []
est1 =  Estudiante("Pablo", 21, "Masculino", "Universitario")
est1.inscripcion(lista_estudiantes)

for e in lista_estudiantes:
    print(e)
```

```python
# 3008 - Dado
# Definir una clase Dado con un atributo “marca”, que corresponde al resultado de lanzar el dado. La clase Dado debe contener el constructor y los siguientes métodos:
# __init__(self,…): constructor que permite inicializar el atributo de la clase Dado. El atributo debe tener un valor predeterminado igual a 0.
# lanzar_dado(self): obtiene un valor aleatorio entre 1 y 6 y se lo asigna al atributo.
# mostrar_puntos(self): muestra por consola el valor del atributo.
import random

class Dado:
    def __init__(self, marca=0):
        self.marca = marca

    def lanzar_dado(self):
        self.marca = random.randint(1, 6)

    def mostrar_puntos(self):
        print(f"El dado muestra: {self.marca}")

d = Dado()
d.lanzar_dado()
d.mostrar_puntos()
```

```python
# 3009 - Vehículo
# Definir una clase Vehiculo y una clase Coche que herede de la clase Vehiculo. La clase Vehiculo tiene 2 atributos: marca y velocidad inicial,
# con un valor predeterminado = 0.
# La subclase Coche tiene un atributo adicional llamado bocina con un valor predeterminado de “tuuut”.
# La clase Vehiculo debe contener el constructor y los siguientes métodos:
# __init__(self,…): constructor que permite inicializar atributos de la clase Vehiculo.
# acelerar(self,v): aumenta en v la velocidad
# desacelerar(self,v): disminuye en v la velocidad
# mostrar_velocidad(self): muestra la velocidad del vehículo
# La subclase Coche debe contener el constructor y el siguiente método adicional:
# __init__(self,…): constructor que permite inicializar los atributos.
# Tocar_claxon(self): muestra por consola una cadena de caracteres que expresa el sonido de la bocina.

class Vehiculo:
    def __init__(self, marca, velocidad=0):
        self.marca = marca
        self.velocidad = velocidad

    def acelerar(self, v):
        self.velocidad += v

    def desacelerar(self, v):
        self.velocidad -= v
        if self.velocidad < 0:
            self.velocidad = 0  # Evita velocidades negativas (opcional)

    def mostrar_velocidad(self):
        print(f"Velocidad actual del vehículo: {self.velocidad} km/h")


class Coche(Vehiculo):
    def __init__(self, marca, velocidad=0, bocina="tuuut"):
        super().__init__(marca, velocidad)
        self.bocina = bocina

    def Tocar_claxon(self):
        print(self.bocina)

Coche1 = Coche("Seat Ibiza")
Coche1.acelerar(50)
Coche1.mostrar_velocidad()
Coche1.Tocar_claxon()
```

```python
# 3010 - Empleado
# Definir una clase Empleado con 4 atributos, el nombre, función, salario y número de horas trabajadas.
# La clase Empleado debe contener el constructor y los siguientes métodos:
# __init__(self,nombre,función,salario): constructor que permite inicializar los atributos. El número de horas trabajadas se inicia a 0 en el constructor.
# trabajar(self,num_horas): aumenta en num_horas las horas ya trabajadas. Además devuelve una cadena con el número de horas trabajadas.
# info_sueldo(self): devuelve una cadena que muestra el salario del empleado.
# dar_aumento(self,cantidad): incrementa en cantidad el salario del empleado
# info_funcion(self): devuelve una cadena con la función del empleado
class Empleado:
    def __init__(self, nombre, funcion, salario):
        self.nombre = nombre
        self.funcion = funcion
        self.salario = salario
        self.horas_trabajadas = 0

    def trabajar(self, num_horas):
        self.horas_trabajadas += num_horas
        return f"Horas trabajadas: {self.horas_trabajadas}"

    def info_sueldo(self):
        return f"Salario actual: {self.salario}"

    def dar_aumento(self, cantidad):
        self.salario += cantidad

    def info_funcion(self):
        return f"Función del empleado: {self.funcion}"

empleado1 = Empleado("Pablo", "Analista", 1500)
print(empleado1.trabajar(8))
print(empleado1.info_sueldo())
empleado1.dar_aumento(200)
print(empleado1.info_sueldo())
print(empleado1.info_funcion())
```

```python
# 3011 - Cálculo numérico
# Definir una clase Calculo_numerico que permita llevar a cabo varias operaciones numéricas. Esta clase tiene un solo atributo que es un número entero.
# Debe contener el constructor y varios métodos de operaciones numéricas:
# __init__(self numero): constructor que permite inicializar el atributo de la clase.
# calculo_factorial(self): retorna el factorial del atributo
# lista_divisores(self): retorna una lista con los divisores del atributo
# esPrimo(self): muestra por consola si el número es primo o no.
# esPar(self): muestra por consola si el número es par o no.
class Calculo_numerico:
    def __init__(self, numero):
        self.numero = numero

    def calculo_factorial(self):
        factorial = 1
        for i in range(1, self.numero + 1):
            factorial *= i
        return factorial

    def lista_divisores(self):
        divisores = []
        for i in range(1, self.numero + 1):
            if self.numero % i == 0:
                divisores.append(i)
        return divisores

    def esPrimo(self):
        if self.numero < 2:
            print(f"{self.numero} no es primo")
            return
        for i in range(2, int(self.numero**0.5) + 1):
            if self.numero % i == 0:
                print(f"{self.numero} no es primo")
                return
        print(f"{self.numero} es primo")

    def esPar(self):
        if self.numero % 2 == 0:
            print(f"{self.numero} es par")
        else:
            print(f"{self.numero} es impar")

num1 = Calculo_numerico(12)
print(num1.calculo_factorial())
print(num1.lista_divisores())
num1.esPrimo()
num1.esPar()
```

```python
# 3012 - Video - (Herencia múltiple)
# Crear la clase Video y la clase Audio. Luego crea una subclase Media que herede de las clases Video y Audio. Los atributos para cada clase son los siguientes:
# Video: título, duración en minutos y categoría
# Audio: título, nombre del artista
# Media: título, categoría, duración y nombre del artista
# Cada clase tiene su constructor y los siguientes métodos:
# Video:
# __init__(self, …): constructor que permite inicializar el atributo de la clase Video.
# mirar_video(self): muestra por consola una cadena indicando que estás viendo el video.
# detener_video(self): muestra por consola una cadena indicando que has detenido la visualización del video.
# Audio:
# __init__(self, …): constructor que permite inicializar el atributo de la clase Audio.
# escuchar_audio(self): muestra por consola una cadena indicando que estás escuchando el audio.
# detener_audio(self): muestra por consola una cadena indicando que has detenido la audición del audio.
# Media:
# __init__(self, …): constructor que permite inicializar el atributo de la clase Media.
class Video:
    def __init__(self, titulo, duracion, categoria):
        self.titulo = titulo
        self.duracion = duracion
        self.categoria = categoria

    def mirar_video(self):
        print(f"Estás viendo el video: {self.titulo}")

    def detener_video(self):
        print(f"Has detenido el video: {self.titulo}")


class Audio:
    def __init__(self, titulo, artista):
        self.titulo = titulo
        self.artista = artista

    def escuchar_audio(self):
        print(f"Estás escuchando: {self.titulo} de {self.artista}")

    def detener_audio(self):
        print(f"Has detenido el audio: {self.titulo}")


class Media(Video, Audio):
    def __init__(self, titulo, duracion, categoria, artista):
        Video.__init__(self, titulo, duracion, categoria)
        Audio.__init__(self, titulo, artista)

video1 = Media("El libro troll", 10, "Humor", "ElRubiusOMG")
video1.mirar_video()
video1.detener_video()
video1.escuchar_audio()
video1.detener_audio()
```
